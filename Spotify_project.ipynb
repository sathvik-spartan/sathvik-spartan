```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd

client_credentials_manager = SpotifyClientCredentials(client_id = "b222ffce42ff45b2b8a73d4137cf7dc4", client_secret="2b0507b81be141e2be1bab01f82bad32")

sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)

playlist_link = "https://open.spotify.com/playlist/37i9dQZF1E4xH7WmIo3NXg"

playlist_URI = playlist_link.split("/")[-1].split('?')[0]

data = sp.playlist_tracks(playlist_URI)

data['items']

data['items'][0]['track']['album']['id']

data['items'][0]['track']['album']['name']

data['items'][0]['track']['album']['release_date']

data['items'][0]['track']['album']

data['items'][0]['track']['album']['total_tracks']

data['items'][0]['track']['album']['external_urls']['spotify']
```

```python
album_list = []

for row in data['items']:

  album_id = row['track']['album']['id']

  album_name = row['track']['album']['name']

  album_release_date = row['track']['album']['release_date']

  album_total_tracks = row['track']['album']['total_tracks']

  album_url = row['track']['album']['external_urls']['spotify']

  album_element = {'album_id':album_id,'name':album_name,'release_date':album_release_date,'total_tracks':album_total_tracks,'url':album_url}

  album_list.append(album_element)

  print(album_list)
```

```python
artist_list=[]
for row in data['items']:
  for key, value in row.items():
    if key =="track":
      # The key is 'artists', not 'artist'
      for artist in value['artists']:
        artist_dict={'artist_id':artist['id'],'artist_name':artist['name'],'external_url':artist['external_urls']['spotify']} #changed href to external_urls for the url
        artist_list.append(artist_dict)
```

```python
from IPython.core.application import SYSTEM_CONFIG_DIRS
song_list = []
for row in data['items']:
  song_id = row['track']['id']
  song_name = row['track']['name']
  song_duration = row['track']['duration_ms']
  song_url = row['track']['external_urls']['spotify']
  song_popularity = row['track']['popularity']
  song_added = row['added_at']
  album_id = row['track']['album']['id']
  artist_id = row['track']['album']['artists'][0]['id']
  song_element = {'song_id':song_id,'song_name':song_name,'duration_ms':song_duration,'url':song_url,'popularity':song_popularity,'song_added':song_added,'album_id':album_id,'artist_id':artist_id}
  song_list.append(song_element)
```

```python
album_df = pd.DataFrame.from_dict(album_list)

album_df.head()

album_df.info()
```
